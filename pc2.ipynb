```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Project CARS 2 Car Data.xlsx' , sheet_name = 'Nordschleife Data')
df.head()
```

```python
df.drop(['Lap Time (min:s)' , 'Real Time (s)'] ,axis = 1, inplace = True)
df.head()
```

```python
#rename column
df.rename(columns = {'Lap Time (s)':'lap_time'} , inplace = True)
df.head()
```

```python
#No. of cars based on Manufacturer
df['Manufacturer'].value_counts().plot(kind='bar',figsize=(20,10), fontsize = 20)
plt.title('No. of cars based on Manufacturer', fontsize = 30)
plt.xlabel('Manufacturer', fontsize = 30)
plt.ylabel('No. of Cars', fontsize = 30 )
plt.show()
```

# Wordcloud on number of cars by Manufacturer


```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
print ('Wordcloud is installed and imported!')
```

```python
word_cloud_string = ''
for col in df['Manufacturer']:
    col = col.replace(" ", "") #replace() doesnt modify og string so you have to equate it!!
    word_cloud_string = word_cloud_string+col+' '
    #print(col)
print(word_cloud_string)
```

```python
# create the word cloud
wordcloud = WordCloud(background_color='white').generate(word_cloud_string)

print('Word cloud created!')
```

```python
#mask
car_mask = np.array(Image.open('car_clip.png'))
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(car_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
acar_wc = WordCloud(width = 1920, height = 1080 , background_color='white', max_words=2000, mask=car_mask)

# generate the word cloud
acar_wc.generate(word_cloud_string)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(acar_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
